## Подготовка к выполнению домашего задания

Подключаем библиотеки:

```python
import plotly.graph_objects as go
import plotly.express as px
import numpy as np
import pandas as pd
```

Выделяем нужные данные:

```python
data_stocks = px.data.stocks()
data_stocks
```

## Задание

```python
fig = go.FigureWidget() # создаем интерактивный рисунок

# график изменения цен акций
fig.add_scatter\
(
    x=data_stocks["date"],
    y=data_stocks["NFLX"],
    mode="markers",
    name="NFLX price"
)

# график изменения стоимости
fig.add_bar\
(
    x=data_stocks["date"],
    y=data_stocks["NFLX"].diff(),
    marker_color=np.where(data_stocks["NFLX"].diff() > 0, 'green', 'red'),
    marker_opacity=0.5,
    name="price difference",
)

# добавляем названия
fig.update_layout(title="Netflix stock price", xaxis_title="date", yaxis_title="relative price")

fig.show()
```
