## Логическая индексация

Начнем с простого. Оставим те строки, где количество сезонов равно трем.

```python
df[df['number_of_seasons'] == 3]
```

Такая форма записи называется логической индексацией.

```python
# теперь оставим те строки где number_of_seasons больше 3
df[df['number_of_seasons'] >= 3]
```

Если нам нужно несколько условий, то применяются базовые логичсекие операции И, ИЛИ и НЕ.  
И = and = &    
ИЛИ = or = |  
НЕ = not = ~  

```python
# нужно учесть что в логической индексации нельзя писать словами
df[df['number_of_seasons'] >= 3 and df['number_of_episodes'] > 10]
```

```python
#  нужно писать через знаки и со скобочками вокргу каждого выражения, иначе будет снова ошибка
df[df['number_of_seasons'] >= 3 & df['number_of_episodes'] < 10]
```

```python
# верная форма записи
df[(df['number_of_seasons'] >= 3) & (df['number_of_episodes'] < 30)]
```

```python
# условий можно сделать столько, сколько нужно
df[((df['number_of_seasons'] >= 3)  & (df['number_of_episodes'] < 30) & (df['name'] == "Black Mirror")) | ( df['number_of_episodes'] > 45) ]
```

Если условий много, то такую запись тяжело читать. Поэтому лучше разбить ее на строки для лучшего вида.

```python
df[((df['number_of_seasons'] >= 3)
     & (df['number_of_episodes'] < 30)
     & (df['name'] == "Black Mirror"))
    | ( df['number_of_episodes'] > 45) ]
```

```python
# индексация не поддерживает двойные сравнения
df[ 30 < df['number_of_episodes'] < 45 ]
```

Првоерки можно делать не только по числам, но и по строкам и по дате.

### query

Есть еще один способ фильтровать данные - с помощью метода query. Внутри него уже можно писать and or и проводить двойные сравнения.

```python
df.query('(number_of_seasons >= 3 and name == "Black Mirror") or number_of_episodes > 40')
```

Обратите внимание: все выражение обернуто в одинарные кавычки, а Black Mirror в двойные.

```python
# двойное сравнение
df.query('30 < number_of_episodes < 40')
```

```python
# проверка на пропуски
df.query('first_air_date_dt.notna()')
```

Еще раз взглянем на срез данных. Обратите внимание на индексы слева - у них сбилась нумерация.

```python
df.query('30 < number_of_episodes < 40')
```

```python
# если мы попробуем прочитать строку с 0 инексом, то получим ошибку
df.query('30 < number_of_episodes < 40').loc[0]
```

```python
#  индекс 2 остался в срезе, поэтому его прочитать можно
df.query('30 < number_of_episodes < 40').loc[2]
```

```python
# чтобы не мучаться с унмерацией применяется iloc - он обращается к индексам в порядке возрастания числа
df.query('30 < number_of_episodes < 40').iloc[0]
```

```python
# прочитаем 15 строку сверху
df.query('30 < number_of_episodes < 40').iloc[15]
```

query позволяет проводить сравннеия со списком значений.

```python
languages_list = ['English', "Español", "Russian"]
```

```python
df.query('spoken_languages in @languages_list')
```

Если мы хотим подать внутрь query внешнюю переменную, то необходимо указать ее через @ , иначе будет ошибка.

```python
number = 4
```

```python
#  неверный вариант
df.query('number_of_seasons == number')
```

```python
#  верный вариант
df.query('number_of_seasons == @number')
```
