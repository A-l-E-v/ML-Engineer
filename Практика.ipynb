```python
means = data.cumsum(axis = 1) # cumsum для всех выборок
means
```

```python
means = means / np.arange(1, 101, 1)
means
```

```python
# @title Графики
plt.figure(figsize=(20, 15))

for i, row in enumerate(means):
    plt.subplot(5, 2, i+1)
    plt.plot(np.arange(1, 101, 1), row, color = 'c', lw = 3)
    plt.hlines(row[-1], 0, 100, color = 'black', lw = 1)
    plt.xlim((-5, 100))

plt.tight_layout()
```
