```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

## Построение 2D-диаграммы

```python
# сгенерируем данные тепла на пластине

data = np.array([[50-((i-50)**2 + (j-50)**2)/100 for i in range(101)] for j in range(101)])
```

```python
data
```

```python
# строим картинку

fig, ax = plt.subplots(figsize=(10, 10))
ax.imshow(data)
plt.show()
```

```python
# добавляем название

fig, ax = plt.subplots(figsize=(10, 10))
ax.set_title("Temperature distribution", size=20)
ax.imshow(data)
plt.show()
```

```python
# добавляем шкалу

fig, ax = plt.subplots(figsize=(10, 10))
ax.set_title("Temperature distribution", size=20)
ax_img = ax.imshow(data)
fig.colorbar(ax_img)
plt.show()
```

```python
# делаем так, что шкала подходит под размеры графика

fig, ax = plt.subplots(figsize=(10, 10))
ax.set_title("Temperature distribution", size=20)
ax_img = ax.imshow(data)
colorbar_pad = 0.04
colorbar_width = 0.06
cax = fig.add_axes([ax.get_position().x1 + colorbar_pad,
                    ax.get_position().y0,
                    colorbar_width,
                    ax.get_position().height])
fig.colorbar(ax_img, cax=cax)
plt.show()
```

```python
# играемся с цветовой гаммой

fig, ax = plt.subplots(figsize=(10, 10))
ax.set_title("Temperature distribution", size=20)
ax_img = ax.imshow(data, cmap='coolwarm')
colorbar_pad = 0.02
colorbar_width = 0.03
cax = fig.add_axes([ax.get_position().x1 + colorbar_pad,
                    ax.get_position().y0,
                    colorbar_width,
                    ax.get_position().height])
fig.colorbar(ax_img, cax=cax)
plt.show()
```

```python
# меняем сглаживание

fig, ax = plt.subplots(figsize=(10, 10))
ax.set_title("Temperature distribution", size=20)
ax_img = ax.imshow(data, cmap='coolwarm', interpolation=None)
colorbar_pad = 0.02
colorbar_width = 0.03
cax = fig.add_axes([ax.get_position().x1 + colorbar_pad,
                    ax.get_position().y0,
                    colorbar_width,
                    ax.get_position().height])
fig.colorbar(ax_img, cax=cax)
plt.show()
```

```python
# прикручиваем контуры

fig, ax = plt.subplots(figsize=(10, 10))
ax.set_title("Temperature distribution", size=20)
ax_img = ax.imshow(data, cmap='coolwarm')
ax.contour(data, levels=10)
# colorbar_pad = 0.02
# colorbar_width = 0.03
# cax = fig.add_axes([ax.get_position().x1 + colorbar_pad,
#                     ax.get_position().y0,
#                     colorbar_width,
#                     ax.get_position().height])
# fig.colorbar(ax_img, cax=cax)
plt.show()
```

```python
# меняем цвет контуров и добавляем подписи

fig, ax = plt.subplots(figsize=(10, 10))
ax.set_title("Temperature distribution", size=20)
ax_img = ax.imshow(data, cmap='coolwarm')
contour = ax.contour(data, levels=10, colors='black')
ax.clabel(contour, fontsize=15)
plt.show()
```

```python
# меняем контур на контурфилл

fig, ax = plt.subplots(figsize=(10, 10))
ax.set_title("Temperature distribution", size=20)
# ax_img = ax.imshow(data, cmap='coolwarm')
contour = ax.contour(data, levels=15, colors='black')
ax.clabel(contour, fontsize=15)
ax.contourf(data, cmap='coolwarm', levels=15)
plt.show()
```
