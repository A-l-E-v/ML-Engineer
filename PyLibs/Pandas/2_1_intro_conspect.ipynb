```python
# Считать эксель файл можно и таким способом
xls = pd.ExcelFile('../data/orig/consulting.xlsx')
sheet = xls.parse('CustomerDemographic')
sheet.head()
```

Если нужно прочитать все страницы в эксель файле, то проще всего это сделать в цикле, сохранив таблицы в список или словарь. Получить полный списко имен таблиц можно через .sheet_names

```python
# прочитаем все таблицы и добавим их в список. Получим список, где каждый элемент - это датафрейм.
df_list = []
for sheet  in xls.sheet_names:
    df = xls.parse(sheet)
    df_list.append(df)
```

```python
# сделаем тоже самое,  но теперь сохраним все в виде словаря, где в качсевте ключей будут имена таблиц
df_dict = {}
for sheet  in xls.sheet_names:
    df = xls.parse(sheet)
    df_dict[sheet] = df
```

## Создание датафрейма "вручную"

Датафрейм/таблицу можно не только прочитать из вне, но и собрать вручную.

```python
# За создание датафрейма отвечает DataFrame()
df = pd.DataFrame()
df
```

Видим, что таблица пуста, поэтому ничего не выводится. Подать данные на вход можно несколькими способами.

```python
# 1. задать таблицу по одному столбцу за раз, подавая на вход массивы с данными
df['number'] = [0,1,2]
df['letter_1'] = ["a", "b", "c"]
df['letter_2'] = ["x", "y", "z"]
df
```

```python
# 2. подать на вход массив с массивами, то есть дать данные по строчно.
# В таком случае надо задать имена столбцов с помощью columns
content = [[0, "a", "x"],
[1, "b", "y"],
[2, "c", "z"]]
pd.DataFrame(content, columns=['number', "letter_1", "letter_2"])
```

```python
# 3. подать на вход массив со словарями. колонки в таком случае задавать не нужно
content = [{'number': 0, "letter_1": "a", "letter_2": "x"} ,
          {'number': 1, "letter_1": "b", "letter_2": "y"},
          {'number': 2, "letter_1": "c", "letter_2": "z"}]
pd.DataFrame(content)
```

## Запись данных

Допустим мы прочитали данные и что то в них изменили. Теперь нам нужно сохранить результат. Это делается методом to_csv, при этом в названии файла надо обязательно указать расширение .csv.

```python
df.to_csv('test_full.csv')
```

В таком виде, как выше, метод сохранит данные в файл в текущей директории в файле с разделителем в виде запятой.  
Если данных много, то процесс записи можнт идти долго. Таблицу можно сохранять не целиком, а только нужные строки или столбцы, что также сократит время записи.

```python
# сохраним первые 50 строк
df.head(50).to_csv('test.csv')
```

Если открыть сохраненные данные, то можно увидеть сто там появился лишний безымянный столбец - это индексы или номера строк. Если они нам не нужны, то можно их удалить, указав index=False.

```python
df.head(50).to_csv('test.csv',  index=False)
```

Как уже было сказано выше, по умолчанию все сохраняется с запятой в роли разделителя.

```python
# вот эти две записи делают идентичную операцию
df.head(50).to_csv('test.csv')
df.head(50).to_csv('test.csv', sep=',')
```

Если мы хотим сменить разделитель, то надо указать параметр sep

```python
# сохраняем данные через точку с запятой
df.head(50).to_csv('test.csv',  index=False, sep=';')
```

```python
# сохраняем данные через табуляцию
df.head(50).to_csv('test.tsv',  index=False, sep='\t')
```

```python
# чтобы сохранить данные в эксель применяется to_excel
df.head(50).to_excel('test.xlsx', index=False)
```

Допустим мы хотим создать эксель файл с некскольими старницами. На первой странице сохраним первые 50 строк, а на второй - последние 50 строк таблицы.

```python
#  данные сохраняются на страницу методом to_excel с укаанием имени страницы
head = df.head(50)
tail = df.tail(50)

writer = pd.ExcelWriter('test.xlsx')
head.to_excel(writer, 'sheet1')
tail.to_excel(writer, 'sheet2')
writer.save()
```

```python
# то же самое можно сделать через меннеджер контекста - так правильнее с токи зрения использования памяти
with pd.ExcelWriter('test2.xlsx') as writer:
    df.head(50).to_excel(writer, 'sheet1')
    df.tail(50).to_excel(writer, 'sheet2')
```
