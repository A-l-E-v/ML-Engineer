## Удалить строки или стобцы

### удалить столбцы

Допустим мы хотим удалить столбец tagline. Это делается методом drop с указанием  axis=1.

```python
# если вызвать метод без указания  axis=1 , т ополучаем ошибку
df.drop('tagline')
```

```python
# правильный вариант
df.drop('tagline', axis=1)
```

```python
# если мы хотим удалить несколько столбцов, то достаточно их перечислить в виде массива
df.drop(['tagline', "original_language"], axis=1)
```

Так же как и в прошлые разы, результат операций нигде не сохраняется, пока мы напрмую это не укажем.

```python
# эти два столбца никуда не делись
df[['tagline', "original_language"]]
```

Если мы хотим удалить их и сохранить результат, то это можно сделать одним из двух способов

```python
# эти две записи делают одно и тоже
df = df.drop(['tagline', "original_language"], axis=1)
df.drop(['tagline', "original_language"], axis=1, inplace=True)
```

```python
# больше столбцов нет
df[['tagline', "original_language"]]
```

### удалить строки

Теперь научимся удалять лишние строки.

```python
df.head()
```

За удаление строк также отвечает метод drop, только теперь на вход ему нужно подавать индексы строк.

```python
# удалим их
df.drop(df.query('number_of_seasons == 8').index)
```

```python
# но проще сделать вот так
df.query('number_of_seasons != 8')
```

На одну операцию меньше, читается проще, а результат тот же самый.

## Дубликаты

Научимся обрабатывать повторы и дубли. За это отвечает метод duplicated, который вовзращает 1, если у ячейки/строки есть дубль.

```python
df.duplicated()
```

```python
# сделаем логическую индексацию
df[df.duplicated()]
```

Если применить duplicated ко всей таблице, то будут искаться полностью совпадающие строки. Видим что таких строк в нашем случае нет.

```python
# если поставить отрицание, то будут найдены уникалные строки. в данном случае - это все строки
df[~df.duplicated()]
```

```python
# если мы хотим найти строки, где потворяются значения в одном солбце, то это делается так
df[df['name'].duplicated()]
```

На вход методу можно подать параметр keep, который может быть равен одному из трех значений:
- last: найти дубли, а оригиналом считать и сохранить первый встреченный элемент
- first: найти дубли, а оригиналом считать и сохранить последний встреченный элемент
- False: найти дубли, и удалить все

```python
# нашли дубли и сохранили первый из них
df[df['name'].duplicated(keep="first")] # last | first | False
```

```python
# нашли дубли и сохранили последний из них
df[df['name'].duplicated(keep="last")] # last | first | False
```

```python
# нашли все дубли
df[df['name'].duplicated(keep=False)] # last | first | False
```

для наглядности отсортируем последний срез по столбцам 'name', 'number_of_seasons' по убыванию

```python
df[df['name'].duplicated(keep=False)].sort_values(['name', 'number_of_seasons'], ascending=False)
```

Корейские названия фильмо анализировать трудно, но видно что имена потовряются

Удаляются дубли методом drop_duplicates с указанием параметра keep (last | first | False)

```python
df.drop_duplicates('name', keep=False)
```

```python
# чтобы сохранить результаты используйте один из двух спообов
df.drop_duplicates('name', keep=False, inplace=True)
df = df.drop_duplicates('name', keep=False)
```

Дубли бывают неявными. Примеры:  
* Game of Thrones - game of thrones  
* 8 - 8.0 - "8"  
* Иван Иванов - Иванов Иван  

В зависимости от задачи и условий, такие дубли надо обрабаывать по своему.
