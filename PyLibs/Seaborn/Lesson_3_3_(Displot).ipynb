```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import matplotlib.ticker as mt
sns.set_theme(style="darkgrid")
```

## 1. Построение гистограмм (histplot)

```python
data_scatter = sns.load_dataset('mpg')
```

```python
data_scatter.head()
```

```python
fig, ax = plt.subplots()
sns.histplot(data=data_scatter, x='mpg', ax=ax)
ax.set_title('Fuel consumption distribution', size=15)
plt.show()
```

```python
# поворачиваем график
fig, ax = plt.subplots()
sns.histplot(data=data_scatter, y='mpg', ax=ax)
ax.set_title('Fuel consumption distribution', size=15)
plt.show()
```

```python
# играемся с бинами
fig, ax = plt.subplots()
sns.histplot(data=data_scatter, x='mpg', bins=10, binwidth=2, ax=ax)
ax.set_title('Fuel consumption distribution', size=15)
plt.show()
```

```python
# можно добавить kde (сглаженную гистограмму)
fig, ax = plt.subplots()
sns.histplot(data=data_scatter, x='mpg', bins=10, kde=True, line_kws={'ls': '--'}, ax=ax)
ax.set_title('Fuel consumption distribution', size=15)
plt.show()
```

```python
# добавляем классы
fig, ax = plt.subplots()
sns.histplot(data=data_scatter, x='mpg', hue='origin', ax=ax)
ax.set_title('Fuel consumption distribution', size=15)
plt.show()
```

```python
# можно наслоить один на другой
fig, ax = plt.subplots()
sns.histplot(data=data_scatter, x='mpg', hue='origin', multiple='stack', ax=ax)
ax.set_title('Fuel consumption distribution', size=15)
plt.show()
```

```python
# можно оставить только ступенчатую функцию от гистограмм
fig, ax = plt.subplots()
sns.histplot(data=data_scatter, x='mpg', hue='origin', element='step', ax=ax)
ax.set_title('Fuel consumption distribution', size=15)
plt.show()
```

```python
# в случае гистограмм разных масштабов их можно независимо нормализовать
fig, ax = plt.subplots()
sns.histplot(data=data_scatter, x='mpg', hue='origin', element='step',
             stat="density", common_norm=False, ax=ax)
ax.set_title('Fuel consumption distribution', size=15)
plt.show()
```

```python
# можно также вместо вероятности появления выдавать процент (имеет смысл в случае категориальных переменных):
fig, ax = plt.subplots()
sns.histplot(data=data_scatter, x='cylinders', stat="percent", discrete=True, ax=ax)
ax.set_title('Number of cylinders distribution', size=15)
plt.show()
```

```python
# можно также вместо вероятности появления выдавать процент (имеет смысл в случае категориальных переменных):
fig, ax = plt.subplots()
sns.histplot(data=data_scatter, x='cylinders', hue='origin',
             multiple='dodge', discrete=True, shrink=0.9, ax=ax)
ax.set_title('Number of cylinders distribution', size=15)
plt.show()
```

```python
# а ещё гистограмму можно "опустошить"
fig, ax = plt.subplots()
sns.histplot(data=data_scatter, x='mpg', fill=False, ax=ax)
ax.set_title('Fuel consumption distribution', size=15)
plt.show()
```

```python
# и ступенчатую функцию тоже можно "опустошить"
fig, ax = plt.subplots()
sns.histplot(data=data_scatter, x='mpg', element='step', fill=False, ax=ax)
ax.set_title('Fuel consumption distribution', size=15)
plt.show()
```

```python
# а ещё можно делать двумерные гистограммы как heatmap'ы!
fig, ax = plt.subplots()
sns.histplot(data=data_scatter, x='mpg', y='horsepower', ax=ax)
ax.set_title('Horsepower distribution over fuel consumption', size=15)
plt.show()
```

```python
# и их тоже можно делить по цветам
fig, ax = plt.subplots()
sns.histplot(data=data_scatter, x='mpg', y='horsepower', hue='origin', ax=ax)
ax.set_title('Fuel consumption distribution over horsepower', size=15)
plt.show()
```

```python
# можно так делать ещё
fig, ax = plt.subplots()
sns.histplot(data=data_scatter, x='mpg', y='origin', hue='origin', ax=ax)
ax.set_title('Fuel consumption distribution over origin', size=15)
plt.show()
```

```python
# можно распространять параметры как tuple
fig, ax = plt.subplots()
sns.histplot(data=data_scatter, x='mpg', y='cylinders', discrete=(False, True), ax=ax)
ax.set_title('Fuel consumption distribution over the number of cylinders', size=15)
plt.show()
```

```python
# Можно ограничивать снизу и сверху значения для попалания в гистограмму
fig, ax = plt.subplots()
sns.histplot(data=data_scatter, x='mpg', y='horsepower', pthresh=.05, pmax=.9, ax=ax)
ax.set_title('Horsepower distribution over fuel consumption', size=15)
plt.show()
```

```python
# Можно прикрутить колорбар
fig, ax = plt.subplots()
sns.histplot(data=data_scatter, x='mpg', y='horsepower', color='orange', cbar=True, ax=ax)
ax.set_title('Horsepower distribution over fuel consumption', size=15)
plt.show()
```

## 2. Построение сглаженных гистограмм (kdeplot)

```python
fig, ax = plt.subplots()
sns.kdeplot(data=data_scatter, x='mpg', ax=ax)
ax.set_title('Fuel consumption distribution', size=15)
plt.show()
```

```python
# поворачиваем график
fig, ax = plt.subplots()
sns.kdeplot(data=data_scatter, y='mpg', ax=ax)
ax.set_title('Fuel consumption distribution', size=15)
plt.show()
```

```python
# играемся с размером ядра
fig, ax = plt.subplots()
sns.kdeplot(data=data_scatter, x='mpg', bw_adjust=1, ax=ax)
ax.set_title('Fuel consumption distribution', size=15)
plt.show()
```

```python
# добавляем классы
fig, ax = plt.subplots()
sns.kdeplot(data=data_scatter, x='mpg', hue='origin', ax=ax)
ax.set_title('Fuel consumption distribution', size=15)
plt.show()
```

```python
# наслаиваем гистограммы
fig, ax = plt.subplots()
sns.kdeplot(data=data_scatter, x='mpg', hue='origin', multiple='stack', ax=ax)
ax.set_title('Fuel consumption distribution', size=15)
plt.show()
```

```python
# наслаиваем гистограммы c нормированием на каждое значение
fig, ax = plt.subplots()
sns.kdeplot(data=data_scatter, x='mpg', hue='origin', multiple='fill', ax=ax)
ax.set_title('Fuel consumption distribution', size=15)
plt.show()
```

```python
# строим ЭФР
fig, ax = plt.subplots()
sns.kdeplot(data=data_scatter, x='mpg', hue='origin',
            cumulative=True, common_norm=False, common_grid=True, ax=ax)
ax.set_title('Fuel consumption distribution', size=15)
plt.show()
```

```python
# добавляем классы, но теперь категориальные, играемся с палеткой
fig, ax = plt.subplots()
sns.kdeplot(data=data_scatter, x='mpg', hue='cylinders', palette='Wistia', ax=ax)
ax.set_title('Fuel consumption distribution', size=15)
plt.show()
```

```python
# добавляем стиля
fig, ax = plt.subplots()
sns.kdeplot(data=data_scatter, x='mpg', hue='cylinders', palette='Wistia',
            fill=True, common_norm=False, lw=0, alpha=0.5, ax=ax)
ax.set_title('Fuel consumption distribution', size=15)
plt.show()
```

```python
# с двумя измерениями тоже можно сглаживать
fig, ax = plt.subplots()
sns.kdeplot(data=data_scatter, x='mpg', y='horsepower', ax=ax)
ax.set_title('Horsepower distribution over fuel consumption', size=15)
plt.show()
```

```python
# и классы добавлять
fig, ax = plt.subplots()
sns.kdeplot(data=data_scatter, x='mpg', y='horsepower', hue='origin', ax=ax)
ax.set_title('Horsepower distribution over fuel consumption', size=15)
plt.show()
```

```python
# заполним цвета
fig, ax = plt.subplots()
sns.kdeplot(data=data_scatter, x='mpg', y='horsepower', hue='origin', fill=True, alpha=0.5, ax=ax)
ax.set_title('Horsepower distribution over fuel consumption', size=15)
plt.show()
```

```python
# можно контролировать количество уровней и нижнюю границу для визуализации
fig, ax = plt.subplots()
sns.kdeplot(data=data_scatter, x='mpg', y='horsepower', hue='origin', fill=True, levels=5, thresh=0.2,
            alpha=0.7, ax=ax)
ax.set_title('Horsepower distribution over fuel consumption', size=15)
plt.show()
```

```python
# а можно построить красивую картинку с большим уровнем сглаживания

formatter = mt.ScalarFormatter()
formatter.set_powerlimits((0, 0))

fig, ax = plt.subplots()
sns.kdeplot(data=data_scatter, x='mpg', y='horsepower', fill=True,
            thresh=0, levels=100, cmap='afmhot',
            cbar=True, cbar_kws={'format': formatter, 'shrink': 0.9}, ax=ax)
ax.set_title('Horsepower distribution over fuel consumption', size=15)
plt.show()
```

## 3. Построение ступенчатых ЭФР (ecdfplot)

```python
fig, ax = plt.subplots()
sns.ecdfplot(data=data_scatter, x='mpg', ax=ax)
ax.set_title('Fuel consumption distribution', size=15)
plt.show()
```

```python
# классы
fig, ax = plt.subplots()
sns.ecdfplot(data=data_scatter, x='mpg', hue='origin', ax=ax)
ax.set_title('Fuel consumption distribution', size=15)
plt.show()
```

```python
# строим не пропорции, а подсчёты
fig, ax = plt.subplots()
sns.ecdfplot(data=data_scatter, x='mpg', hue='origin', stat='count', ax=ax)
ax.set_title('Fuel consumption distribution', size=15)
plt.show()
```

## 4. Построение штрихов (rugplot)

```python
fig, ax = plt.subplots()
sns.kdeplot(data=data_scatter, x='mpg', ax=ax)
sns.rugplot(data=data_scatter, x='mpg', ax=ax)
ax.set_title('Fuel consumption distribution', size=15)
plt.show()
```

```python
# вдоль обоих осей
fig, ax = plt.subplots()
sns.kdeplot(data=data_scatter, x='mpg', y='horsepower', fill=True, ax=ax)
sns.rugplot(data=data_scatter, x='mpg', y='horsepower', ax=ax)
ax.set_title('Horsepower distribution over fuel consumption', size=15)
plt.show()
```

```python
# тоже можно делить по цветам
fig, ax = plt.subplots()
sns.kdeplot(data=data_scatter, x='mpg', y='horsepower', hue='origin', fill=True, ax=ax)
sns.rugplot(data=data_scatter, x='mpg', y='horsepower', hue='origin', ax=ax)
ax.set_title('Horsepower distribution over fuel consumption', size=15)
plt.show()
```

```python
# можно делать выше
fig, ax = plt.subplots()
sns.kdeplot(data=data_scatter, x='mpg', y='horsepower', fill=True, ax=ax)
sns.rugplot(data=data_scatter, x='mpg', y='horsepower', height=.05, ax=ax)
ax.set_title('Horsepower distribution over fuel consumption', size=15)
plt.show()
```

```python
# можно выносить за график
fig, ax = plt.subplots()
sns.kdeplot(data=data_scatter, x='mpg', y='horsepower', fill=True, ax=ax)
sns.rugplot(data=data_scatter, x='mpg', y='horsepower', height=-.02, clip_on=False, ax=ax)
ax.set_title('Horsepower distribution over fuel consumption', size=15)
plt.show()
```

## 5. Использование displot

```python
dp = sns.displot(data=data_scatter, x='mpg', y='horsepower', color='orange', cbar=True, aspect=1.25)
dp.fig.suptitle('Horsepower distribution over fuel consumption', fontsize=15)
dp.fig.subplots_adjust(top=0.9)
plt.show()
```

```python
dp = sns.displot(data=data_scatter, x='mpg', y='horsepower', hue='origin', fill=True,
                 levels=5, thresh=0.2, kind='kde', aspect=1.25)
dp.fig.suptitle('Horsepower distribution over fuel consumption', fontsize=15)
dp.fig.subplots_adjust(top=0.9)
plt.show()
```

```python
dp = sns.displot(data=data_scatter, x='mpg', hue='origin', stat='count', kind='ecdf', aspect=1.25)
dp.fig.suptitle('Fuel consumption distribution', fontsize=15)
dp.fig.subplots_adjust(top=0.9)
plt.show()
```

```python
dp = sns.displot(data=data_scatter, x='mpg', y='horsepower', fill=True, kind='kde', aspect=1.25,
                 rug=True, rug_kws={'height': -.02, 'clip_on': False})
dp.fig.suptitle('Horsepower distribution over fuel consumption', fontsize=15)
dp.fig.subplots_adjust(top=0.9)
plt.show()
```

```python
jp = sns.jointplot(data=data_scatter, x='mpg', y='horsepower')
jp.fig.suptitle('Horsepower distribution over fuel consumption', fontsize=15)
jp.fig.subplots_adjust(top=0.9)
plt.show()
```

```python
jp = sns.jointplot(data=data_scatter, x='mpg', y='horsepower', hue='origin', kind='kde')
jp.fig.suptitle('Horsepower distribution over fuel consumption', fontsize=15)
jp.fig.subplots_adjust(top=0.9)
plt.show()
```
