## Первый запрос

Попробуем получить через API список всех продуктов в магазине (раздел Get all products).  
За это отвечает метод /products. Чтобы в него "постучаться", необходимо отправить GET-запрос на https://fakestoreapi.com/products.  
Если просто перейти по этой ссылке, браузер отобразит нужный нам результат в json формате. Про этот формат мы поговорим чуть позже.

```python
# Это базовый URL для доступа к Fake Store API
BASE_URL = 'https://fakestoreapi.com'
```

```python
# Формируем URL для получения списка продуктов
# Отправляем GET-запрос к API и выведем ответ в формате JSON
product_url = f"{BASE_URL}/products"
response = requests.get(product_url).json()
response
```

Видим результат в JSON формате. В таком виде данные читать неудобно, поэтому преобразуем их к табличному виду.

```python
# Создаем объект DataFrame из данных, полученных из API-ответа
pd.DataFrame(response)
```

## JSON

JSON расшифровывается как JavaScript Object Notation (англ. «объектная запись JavaScript»). По сути это то же самое, что и обычный питоновский словарь, однако JSON не может содержать функций, переменных или комментариев. JSON начинается либо с фигурных скобок, если содержит один объект, либо с квадратных — если список объектов.

```python
# Получаем первый элемент из списка данных, содержащихся в ответе API
response[0]
```

```python
# Извлекаем цену (price) из данных первого элемента в ответе API
response[0]['price']
```

В Python есть встроенный модуль для работы с данными в формате JSON:   

```python
# Импортируем модуль json для работы с данными в формате JSON
import json
```

```python
# метод json.loads() конвертирует JSON-строку в формате питон-словаря
string_data = '{"name": "Генерал Слокам", "date": "15 июня 1904 года"}'
json_data = json.loads(string_data)
json_data
```

```python
# Извлекаем значение 'name' из словаря json_data
json_data['name']
```

```python
# метод loads поддерживает и списки
string_data = '[{"name": "Генерал Слокам", "date": "15 июня 1904 года"}, {"name": "Каморта", "date": "6 мая 1902 года"}]'
json_data = json.loads(string_data)
json_data
```

```python
# Извлекаем значение 'date' из второй строки списка
json_data[1]['date']
```

Метод json.dumps() наоборот, конвертирует Python словарь в формат JSON-строки. При работе с буквами кириллицы указывают аргумент ensure_ascii=False (что значит «работа и за пределами базовой раскладки»).

```python
# Преобразуем словарь json_data в JSON-строку и выведем результат
new_string_data = json.dumps(json_data, ensure_ascii=False)
new_string_data
```

```python
# с результатом операции можно работать как с обычной строкой
new_string_data[2:10]
```

## Запрос одного элемента

## Параметры

Если мы хотим получить только какую-то конкретную информацию в запросе, и нам не нужны остальные данные, то нам необходимо задать фильтры с помощью параметров. Библиотека requests позволяет передавать их в URL. Для этого по ключевому слову params в GET-запросе передают словарь с параметрами.

```python
# получим только первые 5 продуктов
product_url = f"{BASE_URL}/products"
my_params = { # словарь с параметрами запроса
    'limit' : 5
}

response = requests.get(product_url, params=my_params).json()
response
```

В теле функции можно передавать не только параметры, но и прочие данные. Попробуем послать POST запрос с нашими данными, для этого попробуем добавить на сайт новый продукт.

```python
# Отправляем POST-запрос на URL продуктов для создания нового продукта "cute cat" с указанными данными
product_url = f"{BASE_URL}/products"
data = {
    "title": 'cute cat',
    "price": 100,
    "description": 'lovely lovely cat',
    "image": 'https://proprikol.ru/wp-content/uploads/2019/08/kartinki-nyashnye-kotiki-16.jpg',
    "category": 'animal'
}
repsonse = requests.post(product_url, json=data)
```

```python
#  выведем статус код
repsonse.status_code
```

```python
# выведем ответ сервера
repsonse.json()
```
