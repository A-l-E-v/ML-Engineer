3. За какой(какие) месяц(ы) было наибольшее падение продаж? Чему равно это изменение продаж?
4. За какой(какие) месяц(ы) было наименьшее падение продаж? Чему равно это изменение продаж?
5. На каком месяце продажи вышли на плато?

```python
import numpy as np

x = np.arange(14)
y = np.array([-4, -2, 8, 9, 6, 4, 2.5, 2, 1, 0.75, 0.5, 0.25, 0, 0]) + 4


plt.title('Число проданных яйцерезок по месяцам')
plt.xlabel('Месяц')
plt.ylabel('Продажи, тысячи')
plt.plot(x, y)
plt.plot(x, y, 'o')
plt.yticks(np.arange(max(y)+1))
plt.xticks(x)
plt.grid()
```

    Пик продаж: 6-й месяц (170 шт.).

    Наибольший взлет: 3-й месяц (+40 шт.).

```python
import numpy as np
import matplotlib.pyplot as plt

# Пример данных о продажах (месяцы 1-12)
months = np.arange(1, 13)
sales = np.array([10, 30, 70, 120, 150, 170, 160, 140, 130, 125, 120, 120])

# Производная (разница между соседними месяцами)
derivative = np.diff(sales)

plt.figure(figsize=(10, 5))
plt.subplot(1, 2, 1)
plt.plot(months, sales, 'o-')
plt.title("Продажи яйцерезок")
plt.xlabel("Месяц")
plt.ylabel("Продажи")

plt.subplot(1, 2, 2)
plt.plot(months[:-1], derivative, 'r-o')
plt.axhline(0, color='k', linestyle='--')
plt.title("Производная продаж")
plt.xlabel("Месяц")
plt.ylabel("Изменение продаж")

plt.tight_layout()
plt.show()
```

####Задание 5



Вы покупаете участок земли площадью 10 $км^2$ под завод по переработке пластиковых бутылок. Вы можете выбрать форму участка: от прямоугольника 1 на 10 км до квадрата. В целом для Вас не имеет значения форма. Единственная проблема - забор, чем больше периметр участка, тем больше придется отдать за ограду. Забор стоит 2.5 миллиона рублей за км.  

1. Найдите экстремум функции стоимости забора для Вашего будущего участка.
2. Напишите программу, которая покажет график стоимости забора для участка в зависимости от длины одной из сторон. Нанесите точку экстремума.


Функция стоимости:
Пусть стороны a и b, тогда a⋅b=10
Периметр P=2(a+b)=2(a+10/a)
Стоимость C=2.5P

Экстремум:
Минимизируем P:
dP/da=2(1−10/a^2)=0 ⇒ a=sqrt(10)≈3.16 км

Минимальная стоимость: C≈31.6 млн руб

```python
import numpy as np
import matplotlib.pyplot as plt

# Параметры задачи
S = 10  # площадь участка (км²)
price = 2.5  # стоимость забора (млн руб/км)

# Расчет оптимальной стороны квадрата
a_opt = np.sqrt(S)  # оптимальная сторона (км)
b_opt = S / a_opt  # вторая сторона (км)
P_opt = 2 * (a_opt + b_opt)  # периметр (км)
C_opt = price * P_opt  # минимальная стоимость (млн руб)

# Генерация данных для графика
a = np.linspace(1, 10, 100)
b = S / a
P = 2 * (a + b)
C = price * P

# Построение графика
plt.figure(figsize=(10, 6))
plt.plot(a, C, 'b-', label='Стоимость забора')
plt.plot(a_opt, C_opt, 'ro', label=f'Минимум ({a_opt:.2f}, {C_opt:.2f})')

# Подписи точки минимума
plt.text(a_opt, C_opt+2, f'({a_opt:.2f}, {C_opt:.2f})', 
         ha='center', va='bottom', fontsize=10, color='red')

# Настройки графика
plt.title("Зависимость стоимости забора от длины стороны участка")
plt.xlabel("Длина стороны a (км)")
plt.ylabel("Стоимость (млн руб)")
plt.grid(True)
plt.legend()

plt.show()

# Вывод точных значений в консоль
print(f"Оптимальные параметры:")
print(f"Сторона a: {a_opt:.4f} км")
print(f"Сторона b: {b_opt:.4f} км")
print(f"Минимальная стоимость: {C_opt:.4f} млн руб")
```
