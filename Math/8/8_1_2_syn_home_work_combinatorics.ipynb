#### Задание 2

Теперь попробуйте взломать чей-нибудь телефон.

image.png

Нужно ввести $4$ символа, каждый символ выбирается из $10$-ти возможных(в данном случае цифры). В данном случае повторы допустимы, например $1111$. Предположим символов нужно не $4$, а $k$, а выбирается каждый символ не из $10$-ти возможных вариантов, а $n$.

Сколько разных паролей может быть в телефоне с учетом следующих значений?

1.  $n = 19, k = 7$
1.  $n = 16, k = 7$
1.  $n = 16, k = 9$
1.  $n = 10, k = 6$
1.  $n = 7, k = 7$
1.  $n = 14, k = 9$
1.  $n = 19, k = 4$
1.  $n = 9, k = 4$
1.  $n = 12, k = 7$
1.  $n = 6, k = 8$
