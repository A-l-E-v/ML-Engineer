####Задание 1





Вы открыли мужскую парикмахерскую. Пока у вас 4 клиента, двое из них братья. В базе данных произошла ошибка: имена и фамилии сохранились, но неясно, какому имени принадлежит какая фамилия.

Имена   = $\{Василий, \ Петр, \ Николай, \ Иван\}$ \\
Фамилии = $\{Лысов, \ Ирокезов, \ Патлатов\}$

1. Сколько всего есть сочетаний имен и фамилий? По какой формуле это посчитать?
2. Напишите программу, которая выведет все элементы декартового произведения этих двух множеств.

1. Количество сочетаний=∣A∣×∣B∣, где A — имена, B — фамилии.
В данном случае:
4 (имена)×3 (фамилии)=12 сочетаний


```python
from itertools import product

names = ["Василий", "Петр", "Николай", "Иван"]
surnames = ["Лысов", "Ирокезов", "Патлатов"]

combinations = list(product(names, surnames))
print("Все сочетания имен и фамилий:")
for combo in combinations:
    print(combo[0], combo[1])
```

Вы открыли мужскую парикмахерскую. В вашей парикмахерской есть два вида стрижки. За день к вам пришло 3 клиента, и вы знаете их имена.

Имена: $ \ \ \ N = \{Василий, \ Петр, \ Николай \}$ \\
Стрижки: $H =\{Налысо, \ Ирокез \}$

Ваши стрижки за день представляют из себя функцию $f$, которая отображает элементы множества имен $N$ в множество стрижек $H$.

$$f: F \to H$$

Или другими словами составляют пары из имен и стрижек.

Например,

```python
import matplotlib.pyplot as plt

names = ['Vasiliy', 'Petr', 'Nickolay']
cuts  = ['Nalyso', 'Iroqes']

#кодируем имена и стрижки цифрами
x = [0, 1, 2]
y = [0, 1, 1]

plt.figure(figsize=(5, 3))
plt.title('Соответствия имен и стрижек')
plt.xlabel('Имена')
plt.ylabel('Стрижки')

for i in range(len(x)):
  plt.plot(x[i], y[i], 'o')

plt.xticks(x, names)
plt.yticks(y, [cuts[i] for i in y])
plt.grid()
```

Предположим, возможна ситуация когда у всех 3-х клиентов в этот день была одна стрижка.

1. Сколько всего функций $f$ можно составить? По какой формуле это можно посчитать?
2. Перечислите все возможные варианты стрижек за этот день для всех клиентов. Один из примеров представлен на графике: Василий - налысо, Петр - ирокез, Николай - ирокез.
3. Напишите программу, которая решит второй пункт задания. Так как стрижки всего две, то пригодиться двоичная система счисления. Каждый день можно закодировать двоичным числом. Так же, вам помогут функции [enumerate](https://proglib.io/p/python-enumerate-uproshchaem-cikly-s-pomoshchyu-schetchikov-2020-12-08), [bin](https://docs-python.ru/tutorial/vstroennye-funktsii-interpretatora-python/funktsija-bin/) и [zfill](https://pythonz.net/references/named/string.zfill/).


1. Каждому из 3 клиентов можно назначить одну из 2 стрижек.
Формула: ∣H∣^∣N∣=2^3=8
2.     Все клиенты → "Налысо"

    Василий → "Налысо", Петр → "Налысо", Николай → "Ирокез"

    Василий → "Налысо", Петр → "Ирокез", Николай → "Налысо"

    Василий → "Налысо", Петр → "Ирокез", Николай → "Ирокез"

    Василий → "Ирокез", Петр → "Налысо", Николай → "Налысо"

    Василий → "Ирокез", Петр → "Налысо", Николай → "Ирокез"

    Василий → "Ирокез", Петр → "Ирокез", Николай → "Налысо"

    Все клиенты → "Ирокез"

```python
from itertools import product

names = ["Василий", "Петр", "Николай"]
cuts = ["Налысо", "Ирокез"]

print("Все возможные варианты стрижек:")
for i, combo in enumerate(product(cuts, repeat=len(names))):
    print(f"Вариант {i + 1}:")
    for name, cut in zip(names, combo):
        print(f"{name} → {cut}")
```

Дана функция вида $y=(a(x-b))^2+c$.

1. Найдите значения параметров $a, b,$ и $c$ с помощью графика. Поиграйте в [Desmos](https://www.desmos.com/calculator?lang=ru) с возможными значениями. Как можно решить задачу чисто визуально?
2. Напишите программу, которая построит этот график для $x \in [-1, 2].$



Определение параметра c:

    Поскольку вершина находится в (0.5, 1.0), значение c соответствует y-координате вершины.

    c = 1.0

Определение параметра b:

    Координата x вершины равна b.

    b = 0.5

Определение параметра a:

    Используем точку (0, 2), через которую проходит парабола:

2 = (a(0 - 0.5))² + 1
1 = (a(-0.5))²
1 = 0.25a²
a² = 4
a = ±2

По форме параболы определяем знак:

    Если ветви направлены вверх → a = 2

    Если ветви направлены вниз → a = -2

a = 2, так как ветви параболы направлены вверх.

```python
import numpy as np
import matplotlib.pyplot as plt

# Параметры функции
a = 2
b = 0.5
c = 1.0

# Диапазон значений x
x = np.linspace(-1, 2, 400)

# Вычисление y
y = (a * (x - b))**2 + c

# Построение графика
plt.figure(figsize=(8, 6))
plt.plot(x, y, 'b-', linewidth=2)
plt.title('График функции $y = (a(x-b))^2 + c$')
plt.xlabel('x')
plt.ylabel('y')
plt.grid(True)

# Отметка вершины
plt.plot(b, c, 'ro', label=f'Вершина ({b}, {c})')

# Отметка точки (0, 2)
plt.plot(0, 2, 'go', label='Точка (0, 2)')

plt.legend()
plt.show()
```

Постройте график косинуса, у которого:

*   амплитута колебаний равна $\frac{1}{2}.$
*   смещение вправо равно $\frac{1}{4}.$ У стандартного косинуса пик в нуле. Пик должен сдвинуться ровно на $0.25$ вправо.
*   смещение вниз равно $1.$
*   частота колебаний равна 1, то есть полное колебание совершается $1$ раз за $1$ продвижение по $x$. У стандартного синуса колебание совершается раз в $2 \pi \approx 6.28$ по $x$. Длина полного колебания - это расстояние, которое нужно пройти от пика до пика. Подумайте, во сколько раз его нужно ускорить или замедлить.

1. Поиграйте со значениями в Desmos и найдите точную формулу для этого косинуса.
2. Напишите программу, которая построит этот график, $x \in [-1, 3].$ Нарисуйте сетку, она должна идти с шагом $0.1$.





Формула косинуса с параметрами:
y=1/2cos⁡(2π(x−0.25))−1

    Амплитуда: 1/2​.

    Сдвиг фазы: 0.25 вправо.

    Вертикальный сдвиг: −1.

    Частота: 2π для одного колебания на интервале 1.

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(-1, 3, 400)
y = 0.5 * np.cos(2 * np.pi * (x - 0.25)) - 1

plt.figure(figsize=(10, 5))
plt.plot(x, y)
plt.title("Косинус с параметрами")
plt.xlabel("x")
plt.ylabel("y")
plt.xticks(np.arange(-1, 3.1, 0.1))
plt.yticks(np.arange(-1.5, 0.6, 0.1))
plt.grid()
plt.show()
```
