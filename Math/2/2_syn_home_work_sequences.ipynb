###Задание 1





У вас новый, но быстрорастущий интернет-бизнес. Дана последовательность количества ваших клиентов за каждую неделю от запуска.
$$4, 7, 10, 13, 16, … $$
1. Какой формулой описывается последовательность? \\
2. Чему равен 100-ый элемент последовательности? \\
3. Напишите программу, которая выведет все 100 элементов последовательности. \\

1. Данная последовательность является арифметической прогрессией с первым членом a1=4 и разностью d=3.
   Формула: an​=a1​+(n−1)⋅d=4+(n−1)⋅3=3n+1
2. a100​=3⋅100+1=301

```python
# 3-ий вопрос
a1 = 4
d = 3
n = 100

sequence = [a1 + (i - 1) * d for i in range(1, n + 1)]
print("Последовательность:", sequence)
print("100-ый элемент:", sequence[-1])
```

У вас новый, но очень быстрорастущий интернет-бизнес. Дана последовательность вашей прибыли в долларах за каждую неделю от запуска.

$$2, 6, 18, 54, 162, … $$
1. Какой формулой описывается последовательность? \\
2. Чему равен 15-ый элемент последовательности? \\
3. Как с помощью знака суммы записать сумму первых 5-ти элементов этой последовательности? \\
4. Напишите программу, которая выведет все 15 элементов и посчитает их сумму. \\




1. Данная последовательность является геометрической прогрессией с первым членом b1=2 и знаменателем q=3.
Формула общего члена:
bn=b1⋅q^(n−1)=2⋅3^(n−1)

1. b15​=2⋅3^14=2⋅4782969=9565938
2. ![alt text](<Screenshot from 2025-04-23 05-02-58.png>)

```python
b1 = 2
q = 3
n = 15

sequence = [b1 * (q ** (i - 1)) for i in range(1, n + 1)]
total_sum = sum(sequence)
print("Последовательность:", sequence)
print("15-ый элемент:", sequence[-1])
print("Сумма первых 15 элементов:", total_sum)
```

Вы открыли завод по переработке пластиковых бутылок. К сожалению, до бесконечности перерабатывать пластик не выйдет, потому что его качество падает с каждой переработкой. Дана последовательность вашей прибыли в рублях от каждой последовательной переработки одной бутылки.

$$3, 1.5, 0.75, 0.375, … $$
1. Какой формулой описывается последовательность? \\
2. Чему равен 15-ый элемент последовательности? \\
3. Как записать сумму всех элементов этой последовательности (от 1 до $\infty$) с помощью знака суммы? Чему равна эта сумма?
4. Напишите программу, которая выведет первые 15 элементов и посчитает сумму первых 100 элементов. \\

1. Данная последовательность является геометрической прогрессией с первым членом c1=3 и знаменателем q=0.5.
Формула общего члена:
cn=c1⋅q^(n−1)=3⋅0.5^(n−1)
2. c15​=3⋅0.5^14≈0.000183
3. ![alt text](<Screenshot from 2025-04-23 05-11-30.png>) ![alt text](<Screenshot from 2025-04-23 05-11-47.png>)


```python
c1 = 3
q = 0.5
n = 15
n_sum = 100

sequence = [c1 * (q ** (i - 1)) for i in range(1, n + 1)]
sum_100 = sum([c1 * (q ** (i - 1)) for i in range(1, n_sum + 1)])

print("Первые 15 элементов:", sequence)
print("15-ый элемент:", sequence[-1])
print("Сумма первых 100 элементов:", sum_100)
```

3. Напишите программу, которая посчитает сумму первых 100-та элементов множества натуральных чисел в лоб и по формуле. Убедитесь, что они дают один результат.
4. Испробуйте оба метода в разных ячейка для расчета суммы первых $10^8$ натуральных чисел. Какой метод сработал быстрее?

```python
n = 100

# Способ 1: В лоб
sum_brute = sum(range(1, n + 1))

# Способ 2: По формуле
sum_formula = n * (n + 1) // 2

print("Сумма в лоб:", sum_brute)
print("Сумма по формуле:", sum_formula)
```

```python
n = 10**8

# Способ 1 (медленный)
# sum_brute = sum(range(1, n + 1))  # Закомментировано, чтобы не выполнялось

# Способ 2 (быстрый)
sum_formula = n * (n + 1) // 2

print("Сумма по формуле:", sum_formula)
```
