````python
import numpy as np
import matplotlib.pyplot as plt

# Исходные данные
data = np.array([[ 7,  3,  0, 10, 10,  9, 14,  9,  0,  2, 10, 11, 12, 10, 13, 10,
        15, 12,  6,  9, 10, 19, 10, 10,  0, 11, 11, 10,  2,  6],
       [ 5,  7,  8,  7, 12, 20, 15,  8,  1,  1,  8, 15,  5,  5, 11,  5,
         5, 16,  5,  4,  9, 19,  7,  7,  0,  7,  8,  1,  5,  8],
       [16, 10, 20, 13, 16, 14, 14,  7,  6,  5, 14,  0, 15, 15, 13, 15,
        14, 10, 16, 16, 13, 15, 13, 18,  2, 19, 16, 16, 19, 11],
       [ 2,  4,  2, 17,  7, 19,  2,  3,  8,  1, 12, 11,  2, 19,  6, 13,
         6,  4,  3,  1,  9, 13,  3,  3,  1,  2,  2,  4,  8, 12],
       [18, 11,  4, 20, 19, 20,  7, 11,  9, 17,  6, 18,  0, 17,  0, 20,
        19, 17, 19, 19,  5, 17,  1, 18, 19, 19,  4,  8, 20,  9]])

# 1. Построение графиков дневной нагрузки за каждый месяц
plt.figure(figsize=(15, 10))
for i in range(5):
    plt.subplot(3, 2, i+1)
    plt.plot(data[i], marker='o')
    plt.title(f'Месяц {i+1}')
    plt.xlabel('День')
    plt.ylabel('Среднее количество чашек в час')
    plt.grid()
plt.tight_layout()
plt.show()

# 2. Функция freqs для подсчета частот
def freqs(arr):
    unique_values = []
    counts = []
    
    for val in arr:
        if val in unique_values:
            index = unique_values.index(val)
            counts[index] += 1
        else:
            unique_values.append(val)
            counts.append(1)
    
    # Сортируем по значениям
    sorted_pairs = sorted(zip(unique_values, counts))
    unique_values, counts = zip(*sorted_pairs)
    
    return np.array(unique_values), np.array(counts)

# Проверка функции freqs
values, counts = freqs(data[0])
lib_values, lib_counts = np.unique(data[0], return_counts=True)
print("Разница между моей функцией и np.unique:", 
      np.sum(np.abs(values - lib_values)) + np.sum(np.abs(counts - lib_counts)))

# 3. Средняя нагрузка за каждый месяц с использованием freqs
def mean_from_freqs(arr):
    values, counts = freqs(arr)
    return np.sum(values * counts) / np.sum(counts)

monthly_means = [mean_from_freqs(month) for month in data]
lib_means = [np.mean(month) for month in data]
print("Разница между моим расчетом среднего и np.mean:", 
      np.sum(np.abs(np.array(monthly_means) - np.array(lib_means))))

# 4. Графики нагрузки за каждый день с указанием среднего
plt.figure(figsize=(15, 10))
for i in range(5):
    plt.subplot(3, 2, i+1)
    plt.bar(range(1, 31), data[i], width=0.8)
    mean_val = monthly_means[i]
    plt.title(f'Месяц {i+1}, среднее: {mean_val:.2f}')
    plt.xlabel('День')
    plt.ylabel('Чашек в час')
    plt.grid(axis='y')
plt.tight_layout()
plt.show()

# 5. Построение интервального ряда
def create_interval_series(arr, bins=5, range_min=0, range_max=20):
    bin_edges = np.linspace(range_min, range_max, bins+1)
    bin_edges[-1] = 20.01  # Чтобы включить 20 в последний интервал
    
    counts = np.zeros(bins)
    for val in arr:
        for i in range(bins):
            if bin_edges[i] <= val < bin_edges[i+1]:
                counts[i] += 1
                break
    
    bin_centers = (bin_edges[:-1] + bin_edges[1:]) / 2
    bin_width = bin_edges[1] - bin_edges[0]
    
    return bin_centers, counts, bin_width

plt.figure(figsize=(15, 10))
for i in range(5):
    plt.subplot(3, 2, i+1)
    centers, counts, width = create_interval_series(data[i])
    plt.bar(centers, counts, width=width*0.8, align='center')
    plt.title(f'Месяц {i+1}, среднее: {monthly_means[i]:.2f}')
    plt.xlabel('Интервалы нагрузки')
    plt.ylabel('Количество дней')
    plt.grid(axis='y')
    
    # Проверка с plt.hist
    plt.hist(data[i], bins=5, range=(0, 20), alpha=0.3, color='red')
plt.tight_layout()
plt.show()

# 6. Ответ на вопрос об увольнении бариста
print("""
Ответ на вопрос 6:
Анализируя данные за последние 5 месяцев, можно заметить, что в 3-м и 5-м месяцах 
наблюдается существенно более высокая нагрузка на кофе-машину по сравнению с другими месяцами. 
Особенно выделяется 5-й месяц, где среднее количество чашек в час достигает 13.57, 
что значительно выше, чем в первые два месяца (8.47 и 7.93 соответственно).

Это подтверждает слова бариста о повышенной нагрузке в последнее время. 
Поломка могла произойти из-за износа оборудования при такой интенсивной эксплуатации, 
а не по вине сотрудника.

Рекомендация: не увольнять бариста, но рассмотреть возможность:
1. Обслуживания или замены кофе-машины
2. Введения графика профилактического обслуживания
3. Распределения нагрузки между несколькими аппаратами в пиковые периоды
""")
````

### Результат

Решение для задания предоставляется в виде программного кода на языке Python и ответов на вопросы.

````python
from scipy import stats  # Добавляем импорт библиотеки

# 1. Функции для расчета среднего, моды и медианы интервального ряда
def interval_mean(centers, counts):
    total = np.sum(counts)
    return np.sum(centers * counts) / total

def interval_mode(centers, counts):
    max_index = np.argmax(counts)
    return centers[max_index]

def interval_median(centers, counts):
    total = np.sum(counts)
    cumsum = np.cumsum(counts)
    median_pos = total / 2
    
    # Находим интервал, содержащий медиану
    median_bin = np.searchsorted(cumsum, median_pos, side='right')
    
    if median_bin >= len(centers):
        median_bin = len(centers) - 1
    
    # Простая оценка - центр интервала
    return centers[median_bin]

# 2. Применение функций к каждому месяцу
plt.figure(figsize=(15, 10))
for i in range(5):
    centers, counts, width = create_interval_series(data[i])
    
    # Вычисление метрик
    mean_val = interval_mean(centers, counts)
    mode_val = interval_mode(centers, counts)
    median_val = interval_median(centers, counts)
    
    # Проверка с исходными данными
    lib_mean = np.mean(data[i])
    lib_mode = float(stats.mode(data[i])[0])
    lib_median = np.median(data[i])
    
    print(f"Месяц {i+1}:")
    print(f"  Интервальные метрики - Среднее: {mean_val:.2f}, Мода: {mode_val:.2f}, Медиана: {median_val:.2f}")
    print(f"  Точные метрики - Среднее: {lib_mean:.2f}, Мода: {lib_mode:.2f}, Медиана: {lib_median:.2f}")
    print(f"  Разницы: {abs(mean_val-lib_mean):.2f}, {abs(mode_val-lib_mode):.2f}, {abs(median_val-lib_median):.2f}")
    
    # 4. График с метриками
    plt.subplot(3, 2, i+1)
    plt.bar(centers, counts, width=width*0.8, align='center')
    
    # Линии для метрик
    plt.axvline(mean_val, color='r', linestyle='--', label=f'Среднее: {mean_val:.1f}')
    plt.axvline(mode_val, color='g', linestyle=':', label=f'Мода: {mode_val:.1f}')
    plt.axvline(median_val, color='b', linestyle='-.', label=f'Медиана: {median_val:.1f}')
    
    plt.title(f'Месяц {i+1} с центральными метриками')
    plt.xlabel('Интервалы нагрузки')
    plt.ylabel('Количество дней')
    plt.legend()
    plt.grid(axis='y')
plt.tight_layout()
plt.show()

# 5. Ответ на вопрос о лучшем показателе
print("""
Ответ на вопрос 5:
Из трех центральных показателей для предсказания поломки кофе-машины:

1. Лучший показатель: СРЕДНЕЕ значение нагрузки, так как:
   - Учитывает все значения нагрузки
   - Показывает общий уровень эксплуатации оборудования
   - Прямо коррелирует с износом машины (чем больше средняя нагрузка, тем быстрее износ)

2. Худший показатель: МОДА, так как:
   - Показывает только наиболее часто встречающееся значение
   - Может не отражать пиковые нагрузки, которые наиболее опасны для оборудования
   - Например, если большинство дней нагрузка умеренная, но есть несколько дней с экстремальной нагрузкой, мода это не покажет

3. Медиана занимает промежуточное положение:
   - Лучше моды, так как учитывает распределение значений
   - Но хуже среднего, так как не чувствительна к экстремальным значениям

Для профилактики поломок важно учитывать именно среднюю нагрузку, так как она отражает общий уровень использования оборудования.
""")
````
