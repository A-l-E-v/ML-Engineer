Вы открыли мужскую парикмахерскую. В вашей парикмахерской есть два вида стрижки. За день к вам пришло 3 клиента, и вы знаете их имена.

Имена: $ \ \ \ N = \{Василий, \ Петр, \ Николай \}$ \\
Стрижки: $H =\{Налысо, \ Ирокез \}$

Ваши стрижки за день представляют из себя функцию $f$, которая отображает элементы множества имен $N$ в множество стрижек $H$.

$$f: F \to H$$

Или другими словами составляют пары из имен и стрижек.

Например,

```python
import matplotlib.pyplot as plt

names = ['Vasiliy', 'Petr', 'Nickolay']
cuts  = ['Nalyso', 'Iroqes']

#кодируем имена и стрижки цифрами
x = [0, 1, 2]
y = [0, 1, 1]

plt.figure(figsize=(5, 3))
plt.title('Соответствия имен и стрижек')
plt.xlabel('Имена')
plt.ylabel('Стрижки')

for i in range(len(x)):
  plt.plot(x[i], y[i], 'o')

plt.xticks(x, names)
plt.yticks(y, [cuts[i] for i in y])
plt.grid()
```

Предположим, возможна ситуация когда у всех 3-х клиентов в этот день была одна стрижка.
