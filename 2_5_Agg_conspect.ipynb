# Аггрегации

В ходе этого урока мы научимся делать аггрегирующие операции. Агрегация относится к таким операциям, когда больший набор строк свёртывается в меньший. Типичные агрегатные функции - COUNT, MIN, MAX, SUM и MEAN

```python
import pandas as pd
```

```python
df = pd.read_csv("../data/orig/orig_df.csv")
df
```

## Самые частые методы

Рассмотри самое простое и при этом часто используемое.

```python
# максимум
df['number_of_seasons'].max()
```

```python
# минимум
df['number_of_episodes'].min()
```

Это можно применять не только к числовым столбцам, но и к строковым. В таком случае, сортировка и поиск "максимального" будет проводиться по алвафиту.

## Аггрегации

Теперь перейдем к более сложным вещам. Начнем с value_counts, которая находит количество повторений для каждого элемента в столбце

```python
df['number_of_seasons'].value_counts()
```

Видим, что чаще всего сериалы имеют от 1 до 4 сезонов, что логично. 0 - означает что либо это не сериал, а фильм, либо какой то сбой в данных. В реальной рабочей задаче - это надо было бы дополнительно изучить.

Столбец number_of_seasons стал высутпать в роли индексов, поэтому, если мы попробуем прочесть 0 эелемент, то получим не первую строку, а строку где number_of_seasons равно 0

```python
df['number_of_seasons'].value_counts()[0]
```

```python
# при необходимости индексы можно сбросить
df['number_of_seasons'].value_counts().reset_index()
```

```python
# или даже удалить
df['number_of_seasons'].value_counts().reset_index(drop=True)
```

```python
# сбросим индексы, и переименуем столбцы
cnt = df['number_of_seasons'].value_counts().reset_index()
cnt.columns = ['number_of_seasons', 'count']
cnt
```

Если мы хотим сделать боле есложные вычисления, то необходимо применить связку groupby-agg. Внутри groupby указываем поля, по которым будем проводить аггрегацию, а в agg указываем операции и столбцы, к которым эти операции будут применены.
Разрешенные самые частые операции: sum, max, min, mean, median, nunique, count.  
Давайте посчитаем количество сериалов, в разбивке по количеству сезонов, а также узнаем сколько всего было снято серий для таких категорий.

```python
df.groupby('number_of_seasons').agg(cnt=("id", "count"), my_sum=("number_of_episodes", "sum"))
```

группировать можно и по нескольким столбцам

```python
df.groupby(['origin_country', 'languages']).agg(max_vote_count=("vote_count", "max"), min_vote=("vote_average", "min"))
```

```python
# для удобства результаты можно отсортировать
df.groupby(['networks']).agg(mean_time=("episode_run_time", "mean")).sort_values('mean_time', ascending=False)
```

```python
# а пперед этим еще и отфильтровать
df.query(' "US" in origin_country').groupby(['networks']).agg(mean_time=("episode_run_time", "mean")).sort_values('mean_time', ascending=False)
```

```python
# для удобства чтения длиннные записи лучше разбить по строкам
(df
 .query(' "US" in origin_country')
 .groupby(['networks'])
 .agg(mean_time=("episode_run_time", "mean"))
 .sort_values('mean_time', ascending=False))
```

## Сводные таблицы

Рассмотрим другой способ аггрегации - через сводные таблицы.

```python
df.head()
```

```python
# проведем группировку по статус и типу и найдем сумарное кличетво эпизодов
df.groupby(['status', 'type']).agg(sum_season=("number_of_seasons", "sum"))
```

```python
# а теперь тоже самое, но через pivot_table
df.pivot_table(index='status', columns='type', values='number_of_seasons', aggfunc='sum')
```

pivot_table работает следуюим образом: задаем значения из первой колонки как index, значения из второй - как columns, на пересечении считается функция aggfunc для столбца values.

```python
# тоже самое, если применить метод разветывания unstack к groupby
df.groupby(['status', 'type']).agg(sum_season=("number_of_seasons", "sum")).unstack(0)
```

```python
да# тоже самое в транспонированном виде
df.groupby(['status', 'type']).agg(sum_season=("number_of_seasons", "sum")).unstack(0).T
```

Такой подход не всегда годится для решения задачи.

```python
df.groupby(['networks', 'genres']).agg(sum_season=("number_of_seasons", "sum"))
```

```python
# на выходе получаем таблицу, которую сложно анализировать
df.pivot_table(index='networks', columns='genres', values='number_of_seasons', aggfunc='sum')
```
