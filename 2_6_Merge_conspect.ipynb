# Объединение таблиц

В ходе этого занятия мы с вами научимся объединять таблицы.

```python
import pandas as pd
```

```python
# прочитаем данные из эксель файла, и каждую страницы сохраним в отдельный датафрейм
xls = pd.ExcelFile('../data/orig/consulting.xlsx')
cust_dem = xls.parse('CustomerDemographic')
cust_ad = xls.parse('CustomerAddress')
transactions = xls.parse('Transactions')
```

## Merge

Если у таблиц встречаются столбцы с одинаковыми значениями (чаще всего то столбцы с айди номерами), то по этим столбцам таблицы можно объединить. Виды объединений можно рассмотреть на картинке. Логика объединений в пандасе, такая же как в sql, только с чуть урезанным набором опций.

```python
cust_dem.head()
```

```python
cust_ad.head()
```

```python
transactions.head()
```

Во всех таблицах потовряется столбец customer_id. Объединение можно провести по нему. В пандас. за это отвечает метод merge

```python
cust_ad.merge(cust_dem, on="customer_id", how='left')
```

on=столбец, по которому мерджим, а how=тип объединения, бывает left, inner и right

```python
(cust_ad.merge(cust_dem, on="customer_id", how='left').shape,
 cust_ad.merge(cust_dem, on="customer_id", how='inner').shape,
 cust_ad.merge(cust_dem, on="customer_id", how='right').shape,)
```

```python
cust_ad.head()
```

```python
# объединять таблицы можно по нескольки столбцам, в том числе объединять с самой собой
cust_ad.merge(cust_ad, on=["customer_id", "address"], how='inner', suffixes=["_left", "_right"])
```

Параметр suffixes=["_left", "_right"] отвчает за то, чтобы для дублирующихся столбцов добавить к имени суфикс. по умолчанию вместо них подставялется _x _y

```python
# создадим копию таблицы и переименуем столбец
dup =cust_ad.copy()
dup = dup.rename(columns={'customer_id': 'new_customer_id'})
```

```python
# если у столбцов разные имена, то при мердже их надо указвыать так
cust_ad.merge(dup, how='inner', suffixes=["_left", "_right"], left_on='customer_id', right_on='new_customer_id')
```

## Join

join - это очень похожая операция на merge, но в данном случае объединение будет идти по индексам.

```python
# в случае повторов столбцов, выпадет ошибка
cust_ad.join(dup)
```

```python
# верный вариант
cust_ad.join(dup, lsuffix='_left', rsuffix='_right')
```

чтобы продемонстрировать приницп работы объединния по индексам, изеним поряок этих индексов с помщью сортировки и сбросим их.

```python
dup.sort_values('address').reset_index()
```

```python
# теперь снова объеденим
cust_ad.join(dup.sort_values('address').reset_index(), lsuffix='_left', rsuffix='_right')
```

чаще всего join применяется для объединения результатов аггрегации.

```python
x = cust_ad.groupby('state').agg(sum_post=("postcode", "sum"))
x
```

```python
y = cust_ad.groupby('state').agg(mean_pr=("property_valuation", "mean"))
y
```

```python
x.join(y)
```
