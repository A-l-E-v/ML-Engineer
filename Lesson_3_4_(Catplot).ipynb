```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.set_theme(style="darkgrid")
```

## 1. Построение облака точек (stripplot)

```python
data_scatter = sns.load_dataset('mpg')
```

```python
data_scatter.head()
```

```python
fig, ax = plt.subplots()
sns.stripplot(data=data_scatter, x='mpg', ax=ax)
ax.set_title('Fuel consumption distribution', size=15)
plt.show()
```

```python
# добавляем вторую величину
fig, ax = plt.subplots()
sns.stripplot(data=data_scatter, x='mpg', y='origin', ax=ax)
ax.set_title('Fuel consumption distribution by origin', size=15)
plt.show()
```

```python
# меняем ориентацию
fig, ax = plt.subplots()
sns.stripplot(data=data_scatter, y='mpg', x='origin', ax=ax)
ax.set_title('Fuel consumption distribution by origin', size=15)
plt.show()
```

```python
# добавляем цвета
fig, ax = plt.subplots()
sns.stripplot(data=data_scatter, x='mpg', y='origin', hue='origin', ax=ax)
ax.set_title('Fuel consumption distribution by origin', size=15)
plt.show()
```

```python
# можно делить по цветам и иначе
fig, ax = plt.subplots()
sns.stripplot(data=data_scatter, x='mpg', y='origin', hue='cylinders', ax=ax)
ax.set_title('Fuel consumption distribution by origin', size=15)
plt.show()
```

```python
# можно самому подобрать цветовую шкалу
fig, ax = plt.subplots()
sns.stripplot(data=data_scatter, x='mpg', y='origin', hue='cylinders', palette='Oranges', ax=ax)
ax.set_title('Fuel consumption distribution by origin', size=15)
plt.show()
```

```python
# можно разделить облака
fig, ax = plt.subplots()
sns.stripplot(data=data_scatter, x='mpg', y='origin', hue='cylinders', dodge=True, ax=ax)
ax.set_title('Fuel consumption distribution by origin', size=15)
plt.show()
```

```python
# можно сделать облака тонкой линией
fig, ax = plt.subplots()
sns.stripplot(data=data_scatter, x='mpg', y='origin', hue='cylinders', dodge=True, jitter=False, ax=ax)
ax.set_title('Fuel consumption distribution by origin', size=15)
plt.show()
```

```python
# визуализация табличных данных
fig, ax = plt.subplots()
sns.stripplot(data=data_scatter[['mpg', 'horsepower', 'acceleration']], ax=ax)
ax.set_title('Automobiles data visualization', size=15)
plt.show()
```

```python
# визуализация табличных данных со сменой ориентации
fig, ax = plt.subplots()
sns.stripplot(data=data_scatter[['mpg', 'horsepower', 'acceleration']], orient='h', ax=ax)
ax.set_title('Automobiles data visualization', size=15)
plt.show()
```

```python
# также этот параметр используется, когда обе величины числовые
fig, ax = plt.subplots()
sns.stripplot(data=data_scatter, x='mpg', y='cylinders', ax=ax)
ax.set_title('Fuel consumption distribution by number of cylinders', size=15)
plt.show()
```

Заметим, что появилась квадратная сетка

```python
# можно отключить сокращение оси y
fig, ax = plt.subplots()
sns.stripplot(data=data_scatter, x='mpg', y='cylinders', orient='h', native_scale=True, ax=ax)
ax.set_title('Fuel consumption distribution by number of cylinders', size=15)
plt.show()
```

```python
# можно кастомизировать
fig, ax = plt.subplots()
sns.stripplot(data=data_scatter, x='mpg', y='origin', jitter=False, s=20, marker='*', alpha=.5, ax=ax)
ax.set_title('Fuel consumption distribution by origin', size=15)
plt.show()
```

```python
# можно расширить сам рисунок
fig, ax = plt.subplots(figsize=(10, 10))
sns.swarmplot(data=data_scatter, x='mpg', y='cylinders', orient='h', ax=ax)
ax.set_title('Fuel consumption distribution by number of cylinders', size=15)
plt.show()
```

## 3. Построение ящиков с усами (boxplot)

```python
# построение единственного ящика (обратить внимание, что мы используем параметр x в качестве параметра данных!)
fig, ax = plt.subplots()
sns.boxplot(x=data_scatter["mpg"], ax=ax)
ax.set_title('Fuel consumption distribution', size=15)
plt.show()
```

```python
# несколько ящиков по классам
fig, ax = plt.subplots()
sns.boxplot(data=data_scatter, x="mpg", y='origin', ax=ax)
ax.set_title('Fuel consumption distribution by origin', size=15)
plt.show()
```

```python
# несколько ящиков по классам с делением по цветам и вертикальной ориентацией
fig, ax = plt.subplots()
sns.boxplot(data=data_scatter, x='cylinders', y='mpg', hue='origin', ax=ax)
ax.set_title('Fuel consumption distribution by number of cylinders', size=15)
plt.show()
```

```python
# убираем заполнение, добавляем небольшой пробел
fig, ax = plt.subplots()
sns.boxplot(data=data_scatter, x='cylinders', y='mpg', hue='origin', fill=False, gap=.2, ax=ax)
ax.set_title('Fuel consumption distribution by number of cylinders', size=15)
plt.show()
```

```python
# не даём сократить ось x
fig, ax = plt.subplots()
sns.boxplot(data=data_scatter, x='cylinders', y='mpg', hue='origin', fill=False, gap=.2, native_scale=True,
            ax=ax)
ax.set_title('Fuel consumption distribution by number of cylinders', size=15)
plt.show()
```

```python
# несколько ящиков по классам с полным покрытием усами
fig, ax = plt.subplots()
sns.boxplot(data=data_scatter, x="mpg", y='origin', whis=(0, 100), ax=ax)
ax.set_title('Fuel consumption distribution by origin', size=15)
plt.show()
```

```python
# делаем ящики уже вполовину
fig, ax = plt.subplots()
sns.boxplot(data=data_scatter, x="mpg", y='origin', whis=(0, 100), width=.5, ax=ax)
ax.set_title('Fuel consumption distribution by origin', size=15)
plt.show()
```

```python
# играемся с цветами и стилями
fig, ax = plt.subplots()
sns.boxplot(data=data_scatter, x="mpg", y='origin', whis=(0, 100), width=.5,
            color='lime', linecolor='crimson', linewidth=2, ax=ax)
ax.set_title('Fuel consumption distribution by origin', size=15)
plt.show()
```

```python
# играемся с цветами и стилями, используя параметры matplotlib
fig, ax = plt.subplots()
sns.boxplot(data=data_scatter, x="mpg", y='origin', width=.5,
            notch=True, showcaps=False,
            flierprops={"marker": "x"},
            boxprops={"facecolor": (0.33, 0.66, 0.33)},
            medianprops={"color": "crimson", "linewidth": 2},
            ax=ax)
ax.set_title('Fuel consumption distribution by origin', size=15)
plt.show()
```

```python
# можно контролировать уровень сглаживания
fig, ax = plt.subplots()
sns.violinplot(data=data_scatter, x="mpg", y='origin', inner='stick', density_norm='count',
               bw_adjust=.3, ax=ax)
ax.set_title('Fuel consumption distribution by origin', size=15)
plt.show()
```

```python
# можно изменять формат подписей категориальных осей
def formatter(name):
    if name == "usa":
        return "american cars"
    if name == "japan":
        return "japanese cars"
    if name == "europe":
        return "EU cars"
    else:
        raise Exception("Wrong label!")

fig, ax = plt.subplots()
sns.violinplot(data=data_scatter, x="mpg", y='origin', formatter=formatter, ax=ax)
ax.set_title('Fuel consumption distribution by origin', size=15)
plt.show()
```

```python
# можно кастомизировать линию

fig, ax = plt.subplots()
sns.violinplot(x=data_scatter["mpg"], linewidth=2, linecolor='lime', ax=ax)
ax.set_title('Fuel consumption distribution', size=15)
plt.show()
```

```python
# можно кастомизировать внутреннюю линию

fig, ax = plt.subplots()
sns.violinplot(x=data_scatter["mpg"], inner_kws={'box_width': 20,
                                                 'whis_width': 5,
                                                 'color': 'lime'},
               ax=ax)
ax.set_title('Fuel consumption distribution', size=15)
plt.show()
```

```python
# можем применять аггрегирующие функции и избавиться от усов
fig, ax = plt.subplots()
sns.barplot(data=data_scatter, x="model_year", y="mpg", estimator="median", errorbar=None, ax=ax)
ax.set_title('Median fuel consumption by model release year', size=15)
plt.show()
```

```python
# можем подписывать столбцы
fig, ax = plt.subplots()
sns.barplot(data=data_scatter, x="model_year", y="mpg", estimator="median", errorbar=None, ax=ax)
ax.set_title('Median fuel consumption by model release year', size=15)
ax.bar_label(ax.containers[0], fontsize=10)
plt.show()
```

```python
# можем делать аннотации
fig, ax = plt.subplots()
sns.barplot(data=data_scatter, x="model_year", y="mpg", estimator="median", errorbar=None, ax=ax)

heights = [rect.get_height() for rect in ax.containers[0]] # вытаскиваем высоты столбцов
markers = ['v' if next < prev \
           else '^' if next > prev \
           else 's' \
           for prev, next in zip(heights[:-1], heights[1:])] # составляем маркеры для аннотации
colors = ['r' if m == 'v' \
          else 'g' if m == '^' \
          else 'c' \
          for m in markers] # подбираем соответствующие им цвета

for i, (height, marker, color) in enumerate(zip(heights[1:], markers, colors)): # рисуем аннотации по одному
    ax.plot(i + 1, height + 3, marker, markersize=10, color=color)

ax.set_title('Median fuel consumption by model release year', size=15)
ax.bar_label(ax.containers[0], fontsize=10)

plt.show()
```

```python
# если оба параметра числовые, надо не забывать использовать orient
fig, ax = plt.subplots()
sns.barplot(data=data_scatter, x="mpg", y="model_year", orient='h', ax=ax)
ax.set_title('Fuel consumption distribution by model release year', size=15)
plt.show()
```

```python
# меняем внешний вид графика с помощью ключевых слов из pyplot
fig, ax = plt.subplots()
sns.barplot(data=data_scatter, x="mpg", y="origin", errorbar=("pi", 100), capsize=.25,
            err_kws={"color": "r", "linewidth": 2.5},
            linewidth=2.5, edgecolor="gray", facecolor=(0, 0, 0, 0),
            ax=ax)
ax.set_title('Fuel consumption distribution by origin', size=15)
plt.show()
```

## 7. Использование catplot

```python
# stripplot по умолчанию
cp = sns.catplot(data=data_scatter, x='mpg', y='origin', hue='cylinders', palette='Oranges', aspect=1.25)
cp.fig.suptitle('Fuel consumption distribution by origin', fontsize=15)
cp.fig.subplots_adjust(top=0.9)
plt.show()
```

```python
# swarmplot с использованием kind
cp = sns.catplot(data=data_scatter, x='mpg', y='origin', hue='cylinders', palette='Oranges', s=18, aspect=1.5,
                 kind='swarm')
cp.fig.suptitle('Fuel consumption distribution by origin', fontsize=15)
cp.fig.subplots_adjust(top=0.9)
plt.show()
```

```python
# boxplot
cp = sns.catplot(data=data_scatter, x="mpg", y='origin', whis=(0, 100), width=.5, aspect=1.25, kind='box')
cp.fig.suptitle('Fuel consumption distribution by origin', fontsize=15, x=0.6)
cp.fig.subplots_adjust(top=0.9)
plt.show()
```

```python
# violinplot
cp = sns.catplot(data=data_scatter, x="mpg", y='origin', split=True, inner='quart', aspect=1.25, kind='violin')
cp.fig.suptitle('Fuel consumption distribution by origin', fontsize=15, x=0.6)
cp.fig.subplots_adjust(top=0.9)
plt.show()
```

```python
# pointplot
cp = sns.catplot(data=data_scatter, x="model_year", y="mpg", hue='origin',
                 markers=['*', 'o', 's'], linestyles=['-', '--', ':'], aspect=1.25, kind='point')
cp.fig.suptitle('Fuel consumption distribution by model release year', fontsize=15)
cp.fig.subplots_adjust(top=0.9)
plt.show()
```

```python
# barplot
cp = sns.catplot(data=data_scatter, x="model_year", y="mpg", estimator="median", errorbar=None,
                 aspect=1.25, kind='bar')
cp.fig.suptitle('Median fuel consumption by model release year', fontsize=15, x=0.55)
cp.fig.subplots_adjust(top=0.9)
plt.show()
```

```python
# можно делить по колонкам так
cp = sns.catplot(data=data_scatter, x='mpg', y='cylinders', hue='cylinders',
                 col='origin', orient='h')
cp.fig.suptitle('Median fuel consumption by number of cylinders', fontsize=15)
cp.fig.subplots_adjust(top=0.85)
plt.show()
```

```python
# и так
cp = sns.catplot(data=data_scatter, x="model_year", y="mpg", col='origin', kind='bar')
cp.fig.suptitle('Median fuel consumption by model release year', fontsize=15)
cp.fig.subplots_adjust(top=0.85)
plt.show()
```
