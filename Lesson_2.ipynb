# Работаем с построением графика

```python
X = [0, 1, 2, 3, 4, 5, 6]
Y = [1, -1, 1, -1, 1, -1, 1]

fig, ax = plt.subplots()
ax.plot(X, Y)
plt.show()
```

```python
# увеличим размер картинки
X = np.linspace(0, 2*np.pi, 100)

fig, ax = plt.subplots(figsize=(10, 10))
ax.plot(np.cos(X), np.sin(X))
plt.show()
```

```python
# добавляем второй график
fig, ax = plt.subplots(figsize=(10, 10))
ax.plot(np.cos(X), np.sin(X))
ax.plot([-1, 1], [-1, 1])
plt.show()
```

```python
# добавляем подписи
fig, ax = plt.subplots(figsize=(10, 10), facecolor='c')
fig.suptitle('Figure title', size=25)
ax.plot(np.cos(X), np.sin(X))
ax.plot([-1, 1], [-1, 1])
ax.set_title('Axes title', size=20)
ax.set_xlabel('Axis X label', size=15)
ax.set_ylabel('Axis Y label', size=15)
plt.show()
```

```python
# добавляем сетку
fig, ax = plt.subplots(figsize=(10, 10))
ax.plot(np.cos(X), np.sin(X))
ax.plot([-1, 1], [-1, 1])
ax.set_title('Axes title', size=20)
ax.set_xlabel('Axis X label', size=15)
ax.set_ylabel('Axis Y label', size=15)
ax.grid()
plt.show()
```

```python
# добавляем легенду
fig, ax = plt.subplots(figsize=(10, 10))
ax.plot(np.cos(X), np.sin(X), label='circle plot')
ax.plot([-1, 1], [-1, 1], label='line plot')
ax.set_title('Axes title', size=20)
ax.set_xlabel('Axis X label', size=15)
ax.set_ylabel('Axis Y label', size=15)
ax.grid()
ax.legend()
plt.show()
```

```python
# Добавляем маркеры и стили
fig, ax = plt.subplots(figsize=(10, 10))
ax.plot(np.cos(X), np.sin(X), label='circle plot', ls='--', lw=0.5, marker='d', markevery=10)
ax.plot([-1, 1], [-1, 1], label='line plot')
ax.set_title('Axes title', size=20)
ax.set_xlabel('Axis X label', size=15)
ax.set_ylabel('Axis Y label', size=15)
ax.grid(ls='--')
ax.legend(prop={'size':15}, loc='upper left')
plt.show()
```

```python
# добавим решения
fig, ax = plt.subplots(figsize=(10, 10))
ax.plot(np.cos(X), np.sin(X), label='circle plot', ls='--', lw=0.5, marker='d', markevery=10)
ax.plot([-1, 1], [-1, 1], label='line plot')
ax.scatter([-np.sqrt(2)/2, np.sqrt(2)/2], [-np.sqrt(2)/2, np.sqrt(2)/2], color='r')
ax.set_title('Axes title', size=20)
ax.set_xlabel('Axis X label', size=15)
ax.set_ylabel('Axis Y label', size=15)
ax.grid(ls='--')
ax.legend(prop={'size':15}, loc='upper left')
plt.show()
```

```python
# добавим аннотацию
fig, ax = plt.subplots(figsize=(10, 10))
ax.plot(np.cos(X), np.sin(X), label='circle plot', ls='--', lw=0.5, marker='d', markevery=10)
ax.plot([-1, 1], [-1, 1], label='line plot')
ax.scatter([-np.sqrt(2)/2, np.sqrt(2)/2], [-np.sqrt(2)/2, np.sqrt(2)/2], color='r')
ax.set_title('Axes title', size=20)
ax.set_xlabel('Axis X label', size=15)
ax.set_ylabel('Axis Y label', size=15)
ax.annotate("Первое решение", (-np.sqrt(2)/2, -np.sqrt(2)/2), xytext=(-np.sqrt(2)/2 + 0.2, -np.sqrt(2)/2),
            arrowprops={'facecolor':'black', 'width': 1})
ax.annotate("Второе решение", (np.sqrt(2)/2, np.sqrt(2)/2), xytext=(np.sqrt(2)/2 - 0.5, np.sqrt(2)/2),
            arrowprops={'facecolor':'black', 'width': 1})
ax.grid(ls='--')
ax.legend(prop={'size':15}, loc='upper left')
plt.show()
```

# Явный метод программирования

```python
# явный метод
plt.figure(figsize=(10, 10))
plt.plot(np.cos(X), np.sin(X), label='circle plot', ls='--', lw=0.5, marker='d', markevery=10)
plt.plot([-1, 1], [-1, 1], label='line plot')
plt.scatter([-np.sqrt(2)/2, np.sqrt(2)/2], [-np.sqrt(2)/2, np.sqrt(2)/2], color='r')
plt.title('Axes title', size=20)
plt.xlabel('Axis X label', size=15)
plt.ylabel('Axis Y label', size=15)
plt.annotate("Первое решение", (-np.sqrt(2)/2, -np.sqrt(2)/2), xytext=(-np.sqrt(2)/2 + 0.2, -np.sqrt(2)/2),
             arrowprops={'facecolor': 'black', 'width': 1})
plt.annotate("Второе решение", (np.sqrt(2)/2, np.sqrt(2)/2), xytext=(np.sqrt(2)/2 - 0.5, np.sqrt(2)/2),
             arrowprops={'facecolor': 'black', 'width': 1})
plt.grid(ls='--')
plt.legend(loc='upper left')
plt.show()
```

# Работа с множеством графиков

```python
# Делаем много графиков сразу
fig, axs = plt.subplots(2, 2)
fig.suptitle('Multifigure demo', size=15)
axs[0, 0].set_facecolor('b')
axs[-1, -1].set_facecolor('r')
plt.show()
```

```python
# Настраиваем сетку графиков
fig, axs = plt.subplots(2, 2)
fig.suptitle('Multifigure demo', size=15)
gs = fig.add_gridspec(2, 2)
ax = fig.add_subplot(gs[:, 0])
ax.set_facecolor('b')
axs[0, 0].axis('off')
axs[1, 0].axis('off')
plt.show()
```
